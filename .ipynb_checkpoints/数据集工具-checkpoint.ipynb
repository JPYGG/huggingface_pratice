### 2 数据集基本操作

```python
dataset = load_from_disk('./data/ChnSentiCorp/')
```

```python
dataset
```

```python
dataset = dataset['train']
```

```python
# 查看数据内容
dataset[1]
```

```python
for i in range(10):
    print(dataset[i])
```

```python
print(dataset['label'][:10])
```

```python
# 排序
sorted_dataset = dataset.sort('label')
```

```python
print(sorted_dataset['label'][:10])
```

```python
print(sorted_dataset['label'][-10:])
```

```python
# 打乱数据
shuffled_dataset = sorted_dataset.shuffle(seed=10)
```

```python
shuffled_dataset['label'][:10]
```

```python
# 数据抽样
dataset.select([0, 10, 20, 30, 40, 50])
```

```python
# 数据过滤
def f(data):
    return data['text'].startswith('非常不错')

dataset.filter(f)
```

```python
# 训练测试集划分
dataset.train_test_split(test_size=0.1)
```

```python
# 数据分桶: 把数据均匀的分成N份
# 分成4分, 取出索引为 0 的那份
dataset.shard(num_shards=4, index=0)
```

```python
# 重命名字段
dataset.rename_column('text', 'text_rename')
```

```python
# 删除字段
dataset.remove_columns(['text'])
```

```python
# 映射函数
def f(data):
    data['text'] = 'My sentence: ' + data['text']
    return data

maped_dataset = dataset.map(f)
```

```python
print(dataset['text'][:5])
```

```python
print(maped_dataset['text'][:5])
```

```python
# 批处理加速
def f(data):
    text = data['text']
    text = ['My sentence: ' + i for i in text]
    data['text'] = text
    return data
```

```python
maped_dataset = dataset.map(function=f,
           batched=True,
           batch_size=1000,
           num_proc=4)
```

```python
print(dataset['text'][0])
print(maped_dataset['text'][0])
```

```python
# 设置数据格式
dataset.set_format(type='torch', columns=['label'], output_all_columns=True)
```

```python
dataset[20]
```
